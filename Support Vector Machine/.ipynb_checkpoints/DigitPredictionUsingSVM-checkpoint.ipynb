```python
import pandas as pd
from sklearn.datasets import load_digits
digit = load_digits()
```

```python
dir(digit)
```

```python
digit.data[0]
```

```python
digit.target[3]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(digit.data, columns=digit.feature_names)
df.head()
```

```python
df['target'] = digit.target
```

```python
df.head()
```

```python
df1 = df[df.target==1]
df1
```

```python
df2 = df[df.target==2]
df2
```

```python
plt.xlabel('pixel_0_4')
plt.ylabel('pixel_0_6')
plt.scatter(df1['pixel_0_4'],df1['pixel_0_6'],color ='green',marker='+')
plt.scatter(df2['pixel_0_4'],df2['pixel_0_6'],color ='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(digit.data, digit.target, train_size=0.8)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model1 = SVC(kernel='rbf',C=2,gamma=0.0001)
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```
