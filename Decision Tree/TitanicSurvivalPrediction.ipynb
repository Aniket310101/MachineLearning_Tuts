```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
input = df[['Pclass','Sex','Age','Fare']]
input
```

```python
target = df['Survived']
target
```

# Handling the NULL values of Age

```python
import math
median_Age = math.floor(input.Age.median())
median_Age
```

```python
input.Age = input.Age.fillna(median_Age)
```

```python
input
```

# LabelEncoding for Sex

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
input['Sex_n'] = le_sex.fit_transform(input['Sex'])
input
```

```python
input_n = input.drop('Sex',axis='columns')
input_n
```

# Split Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_n,target,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

# Train the Model

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3,28,8,1]])
```
